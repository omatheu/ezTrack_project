# Primeiro modelo aplicado aos dados do projeto:

```python
vetorizar = CountVectorizer(lowercase = False, max_features = 50)
bag_of_words = vetorizar.fit_transform(df['comment'])
```

```python
print(bag_of_words.shape)
```

```python
from sklearn.model_selection import train_test_split

treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words, df['sentiment'], random_state = 42)
```

```python
print(treino.shape)
print(teste.shape)
```

```python
from sklearn.linear_model import LogisticRegression

logisticRegression = LogisticRegression(solver='lbfgs')
logisticRegression.fit(treino, classe_treino)
previsao_teste = logisticRegression.predict_proba(teste)
print(previsao_teste)
```

```python
previsao_teste_bol = previsao_teste[:,1] >= 0.5
print(previsao_teste_bol)
```

```python
previsao_teste_int = previsao_teste_bol.astype(int)
print(previsao_teste_int)
```

```python
from sklearn.metrics import accuracy_score

acuracia_teste = accuracy_score(classe_teste, previsao_teste_int)
print(acuracia_teste)
```

# Criando funções para um pipeline de otimização

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopwords = nltk.corpus.stopwords.words('english')
stopwords.append('https')
print(stopwords)
```

```python
phase_processing = list()

for comment in df['comment']:
    word_text = token_espaco.tokenize(comment)
    new_comment = [w for w in word_text if not w.lower() in stopwords]
    for word in word_text:
        if word not in stopwords:
            new_comment.append(word)
    phase_processing.append(' '.join(new_comment))

df['processing_1'] = phase_processing
```

```python
df['comment'][7]

import re

def find_urls(string):
    url_pattern = re.compile(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+|www.(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    urls = re.findall(url_pattern, string)
    return urls

# Test the function
print(find_urls(df['comment'][7]))
```

```python
def nuvem_palavras_neg(texto, coluna_texto):
    texto_negativo = texto.query("sentiment == -1")
    todas_palavras = ' '.join([texto for texto in texto_negativo[coluna_texto]])
    
    # Removendo as stopwords
    todas_palavras = todas_palavras.split()
    todas_palavras = [palavra for palavra in todas_palavras if palavra not in stopwords]
    
    # Encontrando as URLs
    url_find = find_urls(' '.join(todas_palavras))
    
    # Removendo as palavras que fazem parte de uma URL
    todas_palavras = [palavra for palavra in todas_palavras if palavra not in url_find]
    
    todas_palavras = ' '.join(todas_palavras)
    
    nuvem_palavras = WordCloud(width = 1000, height = 700,
                                max_font_size=110, collocations=False,
                                colormap="tab20c", max_words=150
                                ).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()

nuvem_palavras_neg(df, 'processing_1')
```

```python
def nuvem_palavras_neutro(texto, coluna_texto):
    texto_neutro = texto.query("sentiment == 0")
    todas_palavras = ' '.join([texto for texto in texto_neutro[coluna_texto]])

    # Removendo as stopwords
    todas_palavras = todas_palavras.split()
    todas_palavras = [palavra for palavra in todas_palavras if palavra not in stopwords]
    
    # Encontrando as URLs
    url_find = find_urls(' '.join(todas_palavras))
    
    # Removendo as palavras que fazem parte de uma URL
    todas_palavras = [palavra for palavra in todas_palavras if palavra not in url_find]
    
    todas_palavras = ' '.join(todas_palavras)
    
    nuvem_palavras = WordCloud(width = 1000, height = 700,
                                max_font_size=110, collocations=False,
                                colormap="tab20c", max_words=150
                                ).generate(todas_palavras)
    plt.figure(figsize=(10,7))
    plt.imshow(nuvem_palavras, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    
nuvem_palavras_neutro(df_without_uberfiles, "processing_2")   
```

```python
from sklearn.linear_model import LogisticRegression

def classificar_texto(texto, coluna_texto, coluna_classificacao):
    vetorizar = CountVectorizer(lowercase=False, max_features=50)
    bag_of_words = vetorizar.fit_transform(texto[coluna_texto])
    treino, teste, classe_treino, classe_teste = train_test_split(bag_of_words,
                                                                 texto[coluna_classificacao],
                                                                 random_state = 42)
    
    regressao_logistica = LogisticRegression(solver="lbfgs")
    regressao_logistica.fit(treino, classe_treino)
    
    return regressao_logistica.score(teste, classe_teste)

acuracia_teste = classificar_texto(df, "processing_1", "sentiment")
print(f"{acuracia_teste*100}%")
```

```python
acuracia = classificar_texto(df_without_uberfiles, 'processing_2', 'sentiment')
print(f'{acuracia*100}%')
```

```python
import nltk
import seaborn as sns

def pareto(dataset, quantidade):

    palavras_sem_pontuacao = stopwords + pontuacao

    all_words = ' '.join([text for text in dataset['comment'] if text not in palavras_sem_pontuacao])
    token = nltk.tokenize.WhitespaceTokenizer()
    frequencia = nltk.FreqDist(token.tokenize(all_words))

    df_frequencia = pd.DataFrame({"Palavra": list(frequencia.keys()),
                                    "Frequência": list(frequencia.values())})

    df_frequencia = df_frequencia.nlargest(columns = "Frequência", n = quantidade)

    plt.figure(figsize=(12,8))
    ax = sns.barplot(data=df_frequencia, x= "Palavra", y = "Frequência", color = 'blue')
    ax.set(ylabel = "Contagem")
    plt.show()

pareto(df_without_uberfiles, 10)
```
